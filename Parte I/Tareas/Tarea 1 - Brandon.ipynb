## Tratamiento de datos faltantes

```python

# porcentaje de valores faltantes por columna
datos_faltantes = df_casas_tarea.isnull().mean()*100

# mostrar solo columnas con al mneos un valor faltante

datos_faltantes = datos_faltantes[datos_faltantes> 0].sort_values(ascending= False)

print(datos_faltantes)
```

Se realiza el tratamiento de datos para "BsmtCond"

```python
# Verificar el porcentaje de valores faltantes
porcentaje_faltantes = df_casas_tarea["BsmtCond"].isnull().mean() * 100
print(f"Porcentaje de valores faltantes en BsmtCond: {porcentaje_faltantes:.2f}%")

# Si el porcentaje es menor al 5%, eliminar esos registros
if porcentaje_faltantes < 5:
    df_casas_tarea = df_casas_tarea.dropna(subset=["BsmtCond"])
    print("Filas con valores faltantes en BsmtCond eliminadas.")
else:
    print("M치s del 5% de los datos faltan en BsmtCond. No se eliminaron filas.")

# Mostrar el nuevo n칰mero de observaciones
print(f"Nuevo tama침o del DataFrame: {df_casas_tarea.shape}")
```

## An치lisis de la distribuci칩n de la respuesta y relaci칩n de la respuesta con las preditoras.

**1. SalePrice vs Variables Continuas**
Se analizar치n con:

* Correlaci칩n de Pearson

* Gr치fica de dispersi칩n (scatter plot)

Variables continuas:

* LotArea

* GrLivArea

* GarageArea

* LotFrontage

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Lista de variables continuas
continuas = ["LotArea", "GrLivArea", "GarageArea", "LotFrontage"]

# C치lculo de correlaciones y gr치ficas de dispersi칩n
for var in continuas:
    correlacion = df_casas_tarea[[var, "SalePrice"]].corr().iloc[0, 1]
    print(f"Correlaci칩n de {var} con SalePrice: {correlacion:.4f}")
    
    # Gr치fica de dispersi칩n
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df_casas_tarea, x=var, y="SalePrice")
    plt.title(f"SalePrice vs {var}\nCorrelaci칩n: {correlacion:.2f}")
    plt.xlabel(var)
    plt.ylabel("SalePrice")
    plt.tight_layout()
    plt.show()
```

La 칰nica que muestra una correlaci칩n considerable de 0.72 es GrLivArea que es la superficie habitable, por lo que a mayor superficie es m치s cara

**2. SalePrice vs Variables Discretas (tipo conteo)**
Se analizar치n con:

Estas variables son num칠ricas, pero representan conteos o calificaciones discretas, por eso analizamos: Correlaci칩n de Pearson y gr치fica de dispersi칩n para ver las tendencias

    Variables discretas:

* FullBath

* YearBuilt

* OverallCond

```python
# Lista de variables discretas
discretas = ["FullBath", "YearBuilt", "OverallCond"]

# C치lculo de correlaciones y gr치ficas de dispersi칩n
for var in discretas:
    correlacion = df_casas_tarea[[var, "SalePrice"]].corr().iloc[0, 1]
    print(f"Correlaci칩n de {var} con SalePrice: {correlacion:.4f}")
    
    # Gr치fica de dispersi칩n
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df_casas_tarea, x=var, y="SalePrice", alpha=0.5)
    plt.title(f"SalePrice vs {var}\nCorrelaci칩n: {correlacion:.2f}")
    plt.xlabel(var)
    plt.ylabel("SalePrice")
    plt.tight_layout()
    plt.show()
```
